# Greek GEC Benchmark

Run a local benchmark of Greek-capable LLMs for grammatical error correction using the `Korre` dataset and Ollama.

## Before You Run

Make sure:
- VS Code is using the project kernel.
- Ollama is running locally.
- The models you want to compare are already pulled.

Example terminal commands:

```bash
ollama serve
ollama pull qwen2.5:7b-instruct
ollama pull aya-expanse:8b
ollama pull llama3.1:8b
```

````python
from pathlib import Path
import sys

project_root = Path.cwd()
print("Working directory:", project_root)
print("Python executable:", sys.executable)
````

````python
!ollama list
````

````python
from gec_benchmark import benchmark_ollama_models, save_results, load_gec_dataset

MODELS = [
    "qwen2.5:7b-instruct",
    "aya-expanse:8b",
    "llama3.1:8b",
]

SAMPLE_SIZE = 20
RANDOM_STATE = 42
OUTPUT_DIR = "results/gec_ollama"
````

````python
dataset = load_gec_dataset()
print("Dataset size:", len(dataset))
dataset.head()
````

````python
summary, raw = benchmark_ollama_models(
    models=MODELS,
    sample_size=SAMPLE_SIZE,
    random_state=RANDOM_STATE,
)

summary
````

````python
raw[["model", "original_text", "corrected_text", "prediction"]].head(20)
````

````python
save_results(summary, raw, OUTPUT_DIR)
print(f"Saved benchmark outputs to {OUTPUT_DIR}")
````

````python
summary.to_csv(Path(OUTPUT_DIR) / "summary_preview.csv", index=False)
summary
````
