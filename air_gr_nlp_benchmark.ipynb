````python
import pandas as pd
import importlib

gr_data = pd.read_csv('gen-a/data.csv')
gena = importlib.import_module("gen-a.data_wrapper")
````

## Task: Toxicity

````python
toxicity = gena.ZampieriDt(datasets = gr_data)
toxicity_train = toxicity.get('train')
````

````python
toxicity_train.sample()
````

## Task: Gramatical Error Detection/Correction

````python
gec = gena.KorreDt(datasets=gr_data)
gec_test = gec.get('test')
````

````python
gec_test.sample()
````

## Task: Machine Translation

````python
mt = gena.ProkopidisMtDt(datasets=gr_data)
pd.set_option('display.max_colwidth', None)
for lang in mt.target_langs:
  print(f"Language: {lang} ({mt.langs_dict[lang]})")
  display(mt.get(lang, 'train').sample())
````

## Task: Intent Classification

````python
rizou = gena.RizouDt(datasets=gr_data)
rizou_train = rizou.get('train')
rizou_train.sample()
````

## Task: Named Entity Recognition

````python
ner = gena.BarziokasDt(datasets=gr_data)
ner_train = ner.get('test')
ner_train.sample()
````

## Task: Part of Speech Tagging

````python
prokopidis_ud = gena.ProkopidisUdDt(datasets=gr_data)
prokopidis_ud_train = prokopidis_ud.get('train')
prokopidis_ud_train.head()
````

## Task: Topic Classification

````python
papaloukas = gena.PapaloukasDt(datasets=gr_data)
papaloukas_train = papaloukas.get('train')
papaloukas_train.sample()
````
